# Примена и визуелизација на Брза Фуриева Трансформација во процесирање на сигнали, предпроцесирање на аудио податоци и дигитално процесирање на слики

#### Дефинирање на типовите трансформации употребени во овој проект:

##### Фуриева Трансформација
Фуриева Трансформација се однесува на мапирање на сигнал од временски (просторен) домен во негов спектар, кој припаѓа во рамките на фреквенцискиот домен. Временскиот и фреквенцискиот домен се само алтернативни начини за претставување на сигналите, додека Фуриевата трансформација е математичката врска помеѓу нив. 

##### Дискретна Фуриева Трансформација
Како една од најважните алатки во дигиталното процесирање на сигнали, Дискретна Фуриева Трансформација го пресметува спектарот на сигналот со конечно траење. 
ДФТ е дискретна верзија на Фуриевата трансформација која транформира сигнал (или дискретна секвенца) и негово претставување од временски во фреквенциски домен. Најчест начин на кодирање на информациите е во форма на синусоиди кои формираат сигнал. 

##### Брза Фуриева Трансформација:
Брза Фуриева Трансформација е имплементација на ДФТ која дава скоро исти резултати, но е неверојатно поефикасна и многу побрза, што често го намалува значителното време на пресметка. Ова е алгоритам (збир од алгоритми), кој се користи за брза и ефикасна пресметка на ДФТ.


#### Значењето на ДФТ и БФТ:
Дискретна Фуриева Трансформација е алгоритам кој ги трансформира сигналите од временскиот домен во компоненти на фреквенцискиот домен. 
БФТ е сметачки алгоритам кој го намалува времето на пресметување и сложеноста на големите трансформации. 
ДФТ се изведува со временска комплексност О(n^2), додека БФТ ја редуцира истата на О(n*logn), што е оптимално време за процесирање (обработка) на податоци


|БФТ||ДФТ|
|----||-----| 
|Брза Фуриева Трансформација||Дискретна Фуриева Трансформација|
|БФТ претставува побрза везија на ДФТ алгоритамот||ДФТ е дискретна верзија на Фуриевата Трансформација|
|Различни брзи ДФТ техники за пресметување се познати под едно име како БФТ||Алгоритам кој ги трансформира сигналите од временскиот домен во компоненти од фреквенцискиот домен|
|Имплементација на ДФТ||Воспоставува врска помеѓу претставувањето на временскиот и фреквенцискиот домен|
|Примена во множење на целиброеви и полиноми, алгоритми за филтрирање, пресматување на коефициенти на Фуриеви редови||Примена во решавање на апликации со парцијални равенки, корелациски и спектрални анализи, множење на полиноми|

#### Теоретска дефениција на ДФТ:

ДФТ е метод за конвертирање на секвенца од N комплексни броеви x0, x1,..,xN-1 во нова секвенца од N комплексни броеви изразена како 
$$ X_{k}=\sum\limits_{k=0}^{N-1} x_{n}e^{-2\pi ikn/N}$$
за $ 0 ≤ k ≤ N-1 $ 

$X_k$ е комплексен број кој ја кодира амплитудата и фазата на синусоиден бран со фреквенција $k/N$ циклуси по временска единица (овој дел доаѓа од Ојлеровата формула  $$e^{-2\pi ikn/N}=\ cos(2\pi kn/N)+isin(2\pi kn/N))$$

Идејата на пресметувањето на $X_{k}$ е да се најдат коефициентите на приближувањето на сигналот со линеарна комбинација на ваквите бранови.

Инверзната ДФТ се дефинира на следниот начин:
$$ X_{n}=1/N\sum\limits_{k=0}^{N-1} x_{k}e^{2\pi ikn/N}$$
